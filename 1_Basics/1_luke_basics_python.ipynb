```python
def job_posting_summary(posts):
  num_postings = len(posts)
  sum_sal = sum([job["salary"] for job in posts])
  avg_sal = round(sum_sal / num_postings,2)
  unique_location = list(set(job["location"] for job in posts))

  return {"number of postings": num_postings,
          "average salary": avg_sal,
          "unique locations": unique_location}
```

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate = 0.1, symbol = "$",):
    """
    Initiatilize the salary for managing salary computations.

    """
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol

  def __repr__(self):
    return f"{self.symbol}{self.base_salary:,.0f}"

  def calculate_total_salary(self):
    total_salary = self.base_salary * (1 + self.bonus_rate)
    return f"{self.symbol}{total_salary:,.0f}"
```
