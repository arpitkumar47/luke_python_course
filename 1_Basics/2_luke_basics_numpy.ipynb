# NaN

np.nan

```python
type(np.nan)
```

```python
salary_with_nan = np.array([123_124, np.nan, 145_000, 128_000, 110_000, 149_999, np.nan, 135_000, 115_000, 140_000], dtype=float)
salary_with_nan
```

```python
np.mean(salary_with_nan)
```

```python
np.nanmean(salary_with_nan)
```

```python
# Replace Values with NaN

salary_with_nan[salary_with_nan < 130_000] = np.nan

salary_with_nan
```

```python
# where
# Replace values using np.where
# np.where(condition, if condition is true, if condition is false)

salary_with_nan = np.where(salary_with_nan < 120000, 120000, salary_with_nan)
salary_with_nan
```
